```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('/content/gdrive/My Drive/data/AutoInsurSweden.csv')
```

```
x = np.array(df['number_of_claims'])
y = np.array(df['total_payment(thousands of Swedish Kronor)'])
```

```
def predict(x):
    return 3.6*x
y_pred = predict(x)
```

```
# sum ( (pred - y)^2 for all point) / lengthof(pred or y)
def cost(y_pred, y): #mean squared error
    diff = y_pred-y
    c = np.mean(diff**2)
    return c
```

```
plt.xlabel('number_of_claims')
plt.ylabel('total_payment(thousands of Swedish Kronor)')
plt.scatter(df['number_of_claims'], df['total_payment(thousands of Swedish Kronor)'])
plt.plot(x, y_pred, c='red')
print(cost(y_pred, y))
```

```
def predict_predictor(x, y):
    y_mean = np.mean(y)
    x_mean = np.mean(x)
    grad = y_mean/x_mean
    print('Found gradient', grad)
    def predictor(a):
        return grad * a
    return predictor
```

```
predictor = predict_predictor(x, y)
y_pred = predict(x)
plt.xlabel('number_of_claims')
plt.ylabel('total_payment(thousands of Swedish Kronor)')
plt.scatter(df['number_of_claims'], df['total_payment(thousands of Swedish Kronor)'])
plt.plot(x, y_pred, c='red')
print(cost(y_pred, y))
```

```
#y = mx+b
def straight_line(x, m, b):
    return m*x+b
```

```
def random_search(n_searches=30):
    c_lowest = float("inf")
    for i in range(n_searches):
        m = np.random.randn()
        b = np.random.randn()
        prediction = straight_line(x, m, b)  
        c = cost(prediction, y)
        if c<c_lowest:
            c_lowest=c
            best_m = m
            best_b = b
    return best_m, best_b
    #in a loop
        # randomly pick a value for m and b
        # make a preduction using that value for m and b
        # if the prediction has the lowest cost we've seen so far, note them down somethere
    #return the values for m and b which had the lowest cost
```

```
best_m, best_b = random_search(10000)
print('m:', best_m, 'b:', best_b)
y_pred = straight_line(x, best_m, best_b)
plt.xlabel('number_of_claims')
plt.ylabel('total_payment(thousands of Swedish Kronor)')
plt.scatter(df['number_of_claims'], df['total_payment(thousands of Swedish Kronor)'])
plt.plot(x, y_pred, c='red')
print(cost(y_pred, y))
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
df = pd.read_csv('/content/gdrive/My Drive/data/BikeSharingDaily.csv')
```

```
def cost(y_pred, y): #mean squared error
    diff = y_pred-y
    c = np.mean(diff**2)
    return c
```

```
features = [df['temp'], df['hum']] # change the features you feed our model
X = np.stack(features, axis=1)
Y = np.array(df['cnt'])
n_features = len(features)
print(X.shape)
print(Y.shape)
```

```
def gradient_descent(X, Y, n_steps=30, learning_rate=0.003):
    current_m = np.random.randn(n_features)
    current_b = np.random.randn()

    for step in range(n_steps):
        y_pred = np.array([])
        for x_i in X: #iterating over rows of X
            #[m1, m2] * #[x1, x2] =[m1x2, m2x2]
            y_i = sum(current_m*x_i) + current_b
            y_pred = np.append(y_pred, y_i)

        c = cost(y_pred, Y)

        m_grad = np.zeros(n_features) # [0, 0]
        # current_m = [m_1, m_2]
        for f in range(n_features): #for each feature
            for i in range(X.shape[0]):
                m_grad[f] += (y_pred[i]-Y[i])*X[i][f]
            m_grad[f] *= 2/X.shape[0]

        b_grad = 0
        for i in range(X.shape[0]):
            b_grad += y_pred[i]-Y[i]
        b_grad *= 2/Y.shape[0]

        current_m -= learning_rate*m_grad
        current_b -= learning_rate*b_grad
    
    return current_m, current_b
```

```
best_m, best_b = gradient_descent(X, Y, 1000)
print('m:', best_m, 'b:', best_b)
y_pred = np.matmul(X, best_m) + best_b
print(cost(y_pred, Y))
```

```
plt.scatter(df['instant'], y_pred, label="predicted", marker='x')
plt.scatter(df['instant'], df['cnt'], label="True")
plt.xlabel('Day')
plt.legend()
plt.show()
```

```
from copy import deepcopy
def normalize(X):
    xnorm = deepcopy(X)
    for i in range(X.shape[1]):
        this_column = X[:, i]
        my_mean, my_std = np.mean(this_column), np.std(this_column)
        xnorm[:, i] = (this_column-my_mean)/my_std
    return xnorm
X_norm = normalize(X)

y_mean, y_std = np.mean(Y), np.std(Y)
Y_norm = (Y-y_mean)/y_std
```

```
best_m, best_b = gradient_descent(X_norm, Y_norm, 1000)
print('m:', best_m, 'b:', best_b)
y_pred = np.matmul(X, best_m) + best_b
real_y_pred = (y_pred*y_std) + y_mean
print(cost(real_y_pred, Y))
```

```
plt.scatter(df['instant'], real_y_pred, label="predicted", marker='x')
plt.scatter(df['instant'], df['cnt'], label="True")
plt.xlabel('Day')
plt.legend()
plt.show()
```
