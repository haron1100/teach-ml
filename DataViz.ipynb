```
df = pd.read_csv('/content/gdrive/My Drive/data/iris.csv')

print('Dataset before converting classnames')
print(df.head())

classname_to_number = {
    'Iris-setosa':0,
    'Iris-versicolor':1,
    'Iris-virginica':2
}

classnumber_to_name = {
    0:'Iris-setosa',
    1:'Iris-versicolor',
    2:'Iris-virginica'
}

df['class'] = df['class'].map(classname_to_number)

print('Dataset after converting classnames')
print(df.head())

plt.title('Less good way of visualizing classification')
plt.xlabel('number_of_claims')
plt.ylabel('sepal_length(cm)')
plt.scatter(df['sepal_length(cm)'], df['class'])

plt.figure()

plt.title('Better way of visualizing classification')
plt.ylabel('sepal_length(cm)')
plt.ylabel('sepal_width(cm)')
groups = df.groupby('class')
for name, group in groups:
    plt.scatter(group['sepal_length(cm)'], group['sepal_width(cm)'], label=classnumber_to_name[name])
plt.legend()
#

```
